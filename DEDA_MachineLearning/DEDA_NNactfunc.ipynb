```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1. sigmoid
x = np.linspace(-5, 5, 100)
z = 1 / (1 + np.exp(-x))


plt.figure(0)
plt.plot(x, z)
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")
plt.savefig('sigmoid.png', transparent=True)
plt.show()
```

```python
#2. RELU
def rectified(x):
    return max(0.0, x)
# define a series of inputs
series_in = [x for x in range(-5, 5)]
# calculate outputs for our inputs
series_out = [rectified(x) for x in series_in]
# line plot of raw inputs to rectified outputs
plt.figure(1)
plt.plot(series_in, series_out)
plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.savefig('relu.png', transparent=True)
plt.show()
```

```python
#3. tanh
def tanh(x):
    return (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))

series_in = np.linspace(-5, 5, 100)
series_out = [tanh(x) for x in series_in]
plt.figure(2)
plt.plot(series_in, series_out)
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.savefig('tanh.png', transparent=True)
plt.show()
```

```python
#4 Sigmoid
x1 = np.arange(-5, -0.1, 0.001)
x2 = np.arange(0.001, 5, 0.001)
y1 = x1 * 0
y2 = np.sign(x2)
fig, ax = plt.subplots()
sns.lineplot(x = x1, y = y1, color = (72/256, 128/256, 181/256))
sns.lineplot(x = x2, y = y2, color = (72/256, 128/256, 181/256))
plt.scatter(x=0, y=0, color = (72/256, 128/256, 181/256), facecolors = 'none')
plt.scatter(x=0, y=1, color = (72/256, 128/256, 181/256))
plt.xlabel("x")
plt.ylabel("sgn(x)")
plt.savefig("Signum Activation Function.png", transparent = True)
```
