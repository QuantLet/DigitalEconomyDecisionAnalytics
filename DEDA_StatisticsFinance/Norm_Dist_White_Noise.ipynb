```python
import numpy as np
import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(42)

# Generate white noise: 1001 samples from N(0,1)
x_i = np.random.randn(1001)

# Index i from 0 to 1000
i = np.arange(1001)

# Create plot with transparent background
fig, ax = plt.subplots(figsize=(10,6))

# Set transparent background
fig.patch.set_alpha(0.0)
ax.set_facecolor("none")

# Plot points: all blue
ax.scatter(i, x_i, color='steelblue', s=15)

# Axis settings
ax.set_xlabel("i")
ax.set_ylabel("x_i")
ax.set_xticks(np.arange(0, 1001, 200))
ax.set_yticks(np.arange(-3, 4, 1))
ax.set_xlim(-50, 1050)
ax.set_ylim(-4, 4)

# Title
ax.set_title("Normal Distribution White Noise")

# Show plot
plt.show()
```
